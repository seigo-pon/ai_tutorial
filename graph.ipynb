```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 2 * x
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 2 / x
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = np.tan(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = np.arctan(x)
plt.plot(x, y)
plt.show()
```
